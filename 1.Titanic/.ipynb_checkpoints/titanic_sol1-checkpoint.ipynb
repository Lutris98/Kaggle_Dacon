```python
import numpy as np 
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')  # 기본 scheme말고 seaborn scheme 세팅
sns.set(font_scale=2.5)
```

```python
import warnings  # 은근 warning이 잦음
warnings.filterwarnings('ignore')
%matplotlib inline
```

1. Dataset 확인

```python
train_df=pd.read_csv('C:\\Users\\Dongjun\\.kaggle\\titanic\\train.csv')
test_df=pd.read_csv('C:\\Users\\Dongjun\\.kaggle\\titanic\\test.csv')
train_df.head()
```

```python
train_df.describe()
```

1.1 Null data check

```python
for col in train_df.columns :
    msg = 'column: {:>10}\t Percent of Nan value: {:.2f}%'.format(col,100*train_df[col].isnull().sum()/train_df[col].shape[0])
    print(msg)
```

```python
import missingno as msno  #시각화위한 모듈
msno.matrix(train_df.iloc[:,:], figsize=(8,8))
```

```python
msno.bar(train_df.iloc[:,:], figsize=(8,8))
```

```python
msno.bar(test_df.iloc[:,:], figsize=(8,8))
```

1.2 Target label 확인

```python
f, ax = plt.subplots(1,2,figsize=(18,8))
train_df['Survived'].value_counts().plot.pie(explode=[0,0.1], autopct='%1.1f%%',ax=ax[0])  # explode는 중심에서 각 조각떨어진 정도
ax[0].set_title('Pie plot - Survived')
ax[0].set_ylabel('')
sns.countplot('Survived',data=train_df,ax=ax[1])
ax[1].set_title('Count plot - Survived')
plt.show()
```

2. Exploratory data analysis

2.1 Pclass

```python
pd.crosstab(train_df['Pclass'],train_df['Survived']) # plt나 sns가 아니라 # crosstab좀 애용하자
```

```python
f, ax = plt.subplots(1,2,figsize=(18,8))
train_df['Pclass'].value_counts().plot.bar(ax=ax[0])
ax[0].set_title('Number of Passengers by Pclass')
ax[0].set_ylabel('Count')
sns.countplot('Pclass', data=train_df, hue='Survived', ax=ax[1]) # hue로 crosstab과 같은 의미
ax[1].set_title('Pclass : Survived vs Dead')
plt.show()
```

2.2 Sex

```python
pd.crosstab(train_df['Sex'],train_df['Survived'])
```

```python
f, ax = plt.subplots(1,2,figsize=(18,8))
train_df[['Sex','Survived']].groupby(['Sex']).mean().plot.bar(ax=ax[0])
ax[0].set_title('Number of male / female among Survivors')
sns.countplot('Sex',data=train_df, hue='Survived', ax=ax[1])
ax[1].set_title('Sex : Survived vs Dead')
plt.show()
```

2.3 Both Sex and Pclass

```python
sns.factorplot('Pclass','Survived',data=train_df, hue='Sex')
```
